## Variables

```python
test_user_id = 100
```

## STEP1: Loading Dataset
▪ Loading anime.csv and rating.csv into DataFrame

```python
import pandas as pd
anime_data = pd.read_csv('anime.csv').sort_values(ascending=False, by='rating')
#animeDf = animeDf[:5000]
anime_data
```

```python
rating_data = pd.read_csv('rating.csv')
rating_data = rating_data[:1000000]
rating_data
```

## Merge Dataset by anime_id

```python
raw_merged_data = pd.merge(rating_data, anime_data, on='anime_id', sort = True)
raw_merged_data
```

## Group Anime by name and count the rating for each name

```python
#raw_merged_data = raw_merged_data.groupby('name')['rating_x'].count()
#pd.DataFrame(raw_merged_data)
```

```python
rating_anime_matrix = raw_merged_data.pivot_table(index='user_id', columns=['anime_id'], values = 'rating_x', fill_value= -1)
rating_anime_matrix.head(20)
```

```python
anime_rating_matrix = rating_anime_matrix.transpose()
anime_rating_matrix.head()
```

## Identify Anime rated by 1 user

```python
#user_id = 125
anime_rating_user = anime_rating_matrix[test_user_id]
anime_rating_user.sort_values(ascending= False)
```

## Finding Correlations between a Selected user and all Users

```python
similar_users = anime_rating_matrix.corrwith(anime_rating_user)

#format into data frame
sorted_similar_users = pd.DataFrame(similar_users, columns=['correlation']).sort_values(ascending= False, by= 'correlation')
sorted_similar_users
```

## Recommend Anime based on most similar user 
chosen by best 10 correlation value

```python
#most_similar_users = similar_users.sort_values(by='correlation', ascending= False)
most_similar_users = sorted_similar_users[1:11]
most_similar_users
```

```python
#index = row 
most_similar_users_list = most_similar_users.index.values.tolist()
most_similar_users_list
```

```python
# This shows the rated anime by user[0] (highest correlation)
recommendation = raw_merged_data[raw_merged_data['user_id'] == most_similar_users_list[0]]
recommendation.head(10)
```

## Frame Slicing
Condition 1 = similar user

Condition 2 = rating_y > 5

```python
recommendation = raw_merged_data.loc[(raw_merged_data['user_id'] == most_similar_users_list[0]) & (raw_merged_data['rating_x'] > 5),['name','rating_x']]
recommendation
```
