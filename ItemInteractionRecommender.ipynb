## STEP1: Loading Dataset
▪ Loading anime.csv and rating.csv into DataFrame

```python
import pandas as pd
anime_data = pd.read_csv('anime.csv').sort_values(by = 'anime_id', ignore_index= True)
#animeDf = animeDf[:5000]
anime_data
```

```python
rating_data = pd.read_csv('rating.csv').sort_values(by = 'user_id', ignore_index= True)
rating_data = rating_data[:1000000]
rating_data
```

## Merge Dataset by anime_id

```python
raw_merged_data = pd.merge(rating_data, anime_data, on ='anime_id', sort = True)
raw_merged_data
```

```python
rating_anime_matrix = raw_merged_data.pivot_table(index='user_id', columns=['anime_id'], values = 'rating_x', fill_value= -1)
rating_anime_matrix.head(20)
```

```python
anime_rating_matrix = rating_anime_matrix.transpose()
anime_rating_matrix.head()
```

## Identify Users who rated 1 anime

```python
#anime_id = 125
user_ratings = rating_anime_matrix[125]
user_ratings.sort_values(ascending= False)
```

## Finding Correlation Between an selected anime with all animes

```python
similar_animes = rating_anime_matrix.corrwith(user_ratings)
similar_animes
```

## Recommending The Most Simiar Anime

```python
sorted_similar_anime = pd.DataFrame(similar_animes, columns= ['correlation']).sort_values(by='correlation', ascending= False)
sorted_similar_anime
```

## Recommending The Most Similar Popular Animes
To ensure the highly correlated anime is popular

```python
rating_votes = pd.DataFrame(raw_merged_data.groupby('anime_id')['rating_x'].count())
rating_votes
```

```python
similar_animes_ratings = sorted_similar_anime.join(rating_votes['rating_x']).sort_values(by = 'correlation', ascending = False)
similar_animes_ratings
```

```python
#Get the anime that have higher votes (>=200) and higher correlation
similar_popular_animes = sorted_similar_anime.loc[similar_animes_ratings['rating_x']>=200].dropna()
similar_popular_animes = similar_popular_animes[1:]
similar_popular_animes
```

```python
most_similar_popular_animes = similar_popular_animes[:10]
most_similar_popular_animes_list = most_similar_popular_animes.index.to_list()
most_similar_popular_animes_list
```

```python
target_anime = anime_data.loc[anime_data['anime_id'] == 1]
pd.DataFrame(target_anime)
```

```python
result_anime = anime_data.loc[anime_data['anime_id'].isin(most_similar_popular_animes_list)]
result_anime
```
